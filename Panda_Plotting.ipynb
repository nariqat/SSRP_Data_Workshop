```python
#how to filter data, we can trime columns or rows if we want or if they are insignificant 
data_americas_trimmed = data_americas.iloc[0:,1:]
data_americas_trimmed
```

```python
#GDP is above 2000 will say true, if not it will be false 
data_americas_trimmed > 2000
```

```python
# a mask is a key to tell you which data set you want specifically 
# we set what mask is 
mask = data_americas_trimmed > 2000
data_americas_trimmed[mask]
```

```python
# compare the means and you can see the bottome one is larger because it only takes into consideration the gdps that are > 20000
data_americas.describe()
```

```python
data_americas_trimmed[mask].describe()
```

```python
# which values are larger than the overall mean for each year!
mask_higher = data_americas_trimmed > data_americas_trimmed.mean()
data_americas_trimmed[mask_higher]
```

```python
#see which one is over the overall mean
data_americas_trimmed
```

```python
data_americas_trimmed[mask_higher].describe()
```

```python
#Goal: compute a "wealth index"
    # Define wealth index to be the percentage of years
    # for which a country's GDP is greater than the mean 
    
mask_higher = data_americas_trimmed > data_americas_trimmed.mean()
#use a new function called aggregate, allows us to collapse our data with a specific function 
wealth_score = mask_higher.aggregate('sum', axis=1) / len(data_americas_trimmed.columns)
wealth_score
# axis=1 means sum around the rows while axis+0 means sum around the columns
# length of columns we need to use the functions .column

```
